```python
# Drop columns with excessive missing values
diabetic_data.drop(columns=['weight', 'payer_code', 'medical_specialty'], inplace=True)

# Impute missing categorical values
diabetic_data['race'].fillna('Unknown', inplace=True)
diabetic_data['diag_1'].fillna('Missing', inplace=True)
diabetic_data['diag_2'].fillna('Missing', inplace=True)
diabetic_data['diag_3'].fillna('Missing', inplace=True)

# Verify that no missing values remain
missing_values_after = diabetic_data.isnull().sum()
missing_values_after[missing_values_after > 0]
```

```python
# Impute missing categorical values
diabetic_data = diabetic_data.assign(
    race=diabetic_data['race'].fillna('Unknown'),
    diag_1=diabetic_data['diag_1'].fillna('Missing'),
    diag_2=diabetic_data['diag_2'].fillna('Missing'),
    diag_3=diabetic_data['diag_3'].fillna('Missing')
)

# Verify that no missing values remain
missing_values_after = diabetic_data.isnull().sum()
print(missing_values_after[missing_values_after > 0])
```
